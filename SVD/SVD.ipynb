```python
!pip install surprise
from surprise import SVD
from surprise import Dataset, Reader
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise.model_selection import GridSearchCV
from surprise.model_selection import cross_validate
```

```python
# 'listen_count_bin'이 '10-2704'인 행의 'listen_count_bin' 값을 '10'으로 변경
df.loc[df['listen_count_bin'] == '10~2704', 'listen_count_bin'] = '10'
df['listen_count_bin'] = df['listen_count_bin'].astype(int)

# 결과 출력
print(df)
```

```python
unique_users = df['user_id'].nunique()
unique_tracks = df['track_id'].nunique()

print(f'Unique Users: {unique_users}')
print(f'Unique Tracks: {unique_tracks}')
```

```python
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(df[['user_id', 'track_id', 'listen_count_bin']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# SVD 모델 초기화 및 학습
model = SVD(n_factors=64, random_state=42)  # n_factors는 잠재 요인의 수를 지정합니다.
model.fit(trainset)

# 테스트 데이터에 대한 예측
predictions = model.test(testset)

# 성능 평가 (RMSE 등)
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)

print("RMSE:", rmse)
print("MAE:", mae)
```
