```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def generate_similarity_prompt(lyrics1, lyrics2):
    # GPT-2 모델과 토크나이저 로드
    model_name = "gpt2-large"
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # 프롬프트 생성
    prompt = f"Lyrics 1: {lyrics1}. Lyrics 2: {lyrics2}. Calculate the similarity between these two sets of lyrics for float datatype."

    return prompt

def generate_text(prompt, model_name="gpt2-large", max_length=1024, temperature=0.8, top_k=50):
    # GPT-2 모델과 토크나이저 로드
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # 프롬프트를 인코딩
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # 모델을 사용하여 텍스트 생성
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        pad_token_id=tokenizer.eos_token_id  # 문장 끝을 나타내는 토큰 추가
    )

    # 생성된 텍스트 디코딩
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    return generated_text

def calculate_similarity(lyrics1, lyrics2):
    # TF-IDF 벡터화
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([lyrics1, lyrics2])

    # 코사인 유사도 계산
    similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

    return similarity

# 비교할 두 가사
# lyrics1 = "I've got sunshine on a cloudy day."
# lyrics2 = "When it's cold outside, I've got the month of May."

# 프롬프트 생성
prompt = generate_similarity_prompt(lyrics1, lyrics2)

# 생성된 텍스트 출력
generated_text = generate_text(prompt, model_name="gpt2-large")
print(generated_text)

# 가사 유사도 계산
similarity_score = calculate_similarity(lyrics1, lyrics2)
print(f"Lyrics Similarity Score: {similarity_score}")
```

```python
from transformers import BertModel, BertTokenizer
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from transformers import BertTokenizer, BertForSequenceClassification

from sklearn.metrics.pairwise import cosine_similarity
import torch
def get_bert_embedding(text, model, tokenizer):
    # 토큰화 및 패딩
    input_ids = tokenizer.encode(text, return_tensors="pt", padding=True, truncation=True)

    # BERT 모델로부터 임베딩 얻기
    with torch.no_grad():
        outputs = model(input_ids)
    # [CLS] 토큰의 임베딩 얻기
    cls_embedding = outputs.last_hidden_state[:, 0, :]

    return cls_embedding

# BERT 모델 및 토크나이저 로드
# model_name = 'bert-base-uncased'
# model = BertModel.from_pretrained(model_name)
# tokenizer = BertTokenizer.from_pretrained(model_name)
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# # 비교할 두 가사
# lyrics1 = "I've got sunshine on a cloudy day."
# lyrics2 = "When it's cold outside, I've got the month of May."

# 각 가사에 대한 BERT 임베딩 얻기
embedding1 = get_bert_embedding(depression_lyrics[0], model, tokenizer)
embedding2 = get_bert_embedding(exciting_lyrics[1], model, tokenizer)

# 임베딩 간의 코사인 유사도 계산
similarity_score = cosine_similarity(embedding1, embedding2)[0][0]

print(f"BERT Embedding Cosine Similarity Score: {similarity_score}")
```
